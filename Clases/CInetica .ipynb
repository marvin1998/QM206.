import matplotlib.pyplot as plt
import numpy as np

## Cinética de Reacción Consecutiva
## A-->B

```python
A=1
B=0
C=0
k1=0.1
k2=0.1
dt=0.1
t=0
x=[]
yA=[]
yB=[]
yC=[]
while t<60:
    x.append(t)
    yA.append(A)
    yB.append(B)
    yC.append(C)
    dA=-k1*A*dt
    dB=-k2*B*dt+k1*A*dt
    dC=k2*B*dt
    A+=dA
    B+=dB
    C+=dC
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.plot(x,yC)
plt.xlabel("t(min)")
plt.ylabel("Concentración")
plt.title("Cinética de Reacción Consecutiva")
plt.show
```

```python
A=1.0
B=0
k1=0.1
k2=0.2
dt=0.1
t=0
x=[]
yA=[]
yB=[]
while t<60:
    x.append(t)
    yA.append(A)
    yB.append(B)
    dA=-k1*A*A*dt
    dB=+1*A*dt-k2*B*dt
    A+=dA
    B+=dB
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel("t(min)")
plt.ylabel("Concentración")
plt.title("Cinética de Reacción Consecutiva")
plt.show()
```
