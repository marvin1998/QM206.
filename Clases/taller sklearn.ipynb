```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris= datasets.load_iris ()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])
iris_dataframe["target"]=iris.target
x=iris_dataframe["SL"]
print(x.mean())
y=iris_dataframe["SW"]
col=(iris_dataframe["target"]+1)*2
print(y.mean())
PW=iris_dataframe["PW"]*50
```

```python
plt.scatter(x,y,c=col,marker=10,s=PL,alpha=0.5)
plt.xlabel("Largo del Sépalo(cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.title("Ancho de Sépalo vs Largo de Sépalo")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,c=col,s=PW,alpha=1.5)
plt.xlabel("Largo del Sépalo(cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.title("Ancho de Sépalo vs Largo de Sépalo")
plt.grid(True)
plt.show()
```
