```python
plt.plot([0,2,4,1],[3,1,7,4],color="yellowgreen",linewidth=7,alpha=0.9,marker="o",markersize=15,markeredgecolor="g",markerfacecolor="red")
plt.xlabel("$x(cm²)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("EXperimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.title("funcion Sen(4x)0-$2\pi$")
plt.grid(True)
plt.plot(x,y,color="Green")
plt.plot(x,y2,color="darkslategray", dashes=[1,1,2,2])
plt.xlabel("t(seg)")
plt.ylabel("Amplitud de Onda")
plt.show
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*250+50
plt.scatter(x,y,s=r,color="magenta")
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="green",edgecolor="black",density=True)
plt.title("Distribución de la Población")
plt.ylabel("Densidad de habitantes/$m²$")
plt.text(-3,.35,r"$\mu=100,\ \sigma=1.0$")
plt.show()
```
